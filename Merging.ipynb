```python
import pandas as pd
```

```python
sept= pd.read_csv("JC-202009-citibike-tripdata.csv")
sept.head()
```

```python
oct=pd.read_csv("JC-202010-citibike-tripdata.csv")
oct.head()
```

```python
nov=pd.read_csv("JC-202011-citibike-tripdata.csv")
nov.head()
```

```python
tableau_df1=pd.merge(sept, oct, on="tripduration", how="left")
tableau_df1
```

```python
tableau_df1 = tableau_df1.rename(columns={'Oct_start_time':'Sept_start_time', 'Oct_stop_time':'Sept_stop_time',
                                         'Oct_start_station_id':'Sept_start_station_id', 'Oct_start_station_name':
                                         'Sept_start_station_name','Oct_start_lat':'Sept_start_lat',
                                         'Oct_start_long': 'Sept_start_long', 'Oct_end_station_d':
                                         'Sept_end_station_d', 'Oct_end_station_name': 'Sept_end_station_name',
                                         'Oct_end_lat': "Sept_end_lat", 'Oct_end_long':
                                         'Sept_end_long'})
tableau_df1.head()
```

```python
tableau_df1 = tableau_df1.rename(columns={'start station latitude_y':'Oct_start_lat', 'start station longitude_y':
                                          'Oct_start_long', 'end station id_y': 'Oct_end_station_id', 'end station name_y':
                                          'Oct_end_station_name', 'end station latitude_y': 'Oct_end_lat', 'end station longitude_y':
                                          'Oct_end_long', 'bikeid_y': 'bike_id', 'usertype_y': 'usertype', 'birth year_y': 
                                          'birthyear', 'gender_y': 'gender'})
tableau_df1.head()                                 
```

```python
tableau_df2=pd.merge(tableau_df1, nov, on="tripduration", how="left")
tableau_df2.head()
```

```python
tableau_df2 = tableau_df2.rename(columns={'starttime':'Nov_start_time','stoptime':'Nov_stoptime',
                                          'start station id': 'Nov_start_station_id', 'start station name':
                                          'Nov_start_station_name', 'start station latitude':'Nov_start_lat',
                                          'start station longitude':'Nov_start_long', 'end station id':'Nov_end_id',
                                          'end station name':'Nov_end_station_name','end station latitude':'Nov_end_lat',
                                          'end station longitude':'Nov_end_long','usertype_y':'usertype','gender_y':'gender'})
tableau_df2.head()
```

```python
tableau_df2.to_csv("tableau_hw.csv", encoding='utf-8')
```
